# Stacks

```python
stacks = []
choice = 0
while(choice != 5):
  print("1. To push element")
  print("2. To pop element")
  print("3. Check top/peep")
  print("4. Search element")
  print("5. Exit")
  choice = int(input("Enter a operation "))
  if choice == 1:
    ele= int(input("Enter an elemnt "))
    stacks.append(ele)
    print("Pushed successfully")
  elif choice == 2:
    stacks.pop()
    print("Poped successfully")
  elif choice == 3:
    top = stacks[len(stacks)-1]
    print(f"Top Element = {top}")
  elif choice == 4:
    ele= int(input("Enter an elemnt to search "))
    if ele in stacks:
      print("present at position {}".format(stacks.index(ele)))
    else:
      print("Not present")
  else:
    break


  
```

#Queues
FIFO

```python
queue=[]
choice = 0
while(choice != 7):
  print("1. Insert element")
  print("2. Remove element")
  print("3. Search an element")
  print("4. Check if empty")
  print("5. checl front")
  print("6. checl rear")
  print("7. exit")
  choice = int(input("Enter a operation "))
  if choice == 1:
    ele = int(input("Enter an element"))
    queue.append(ele)
    print("-------------------------Element inserted----------------------------------")
  elif choice == 2:
    queue.pop(0)
    print("-------------------------Element removed-----------------------------------")
  elif choice == 3:
    ele = int(input("Enter an element to serach"))
    if ele in queue:
      print("----------Present at position {}---------------------------".format(queue.index(ele)))
    else:
      print("---------------------------Not present---------------------------------")
  elif choice == 4:
    if(len(queue)==0):
      print("-----------------------------Empty-----------------------------------")
    else:
      print("------------------------------Not empty------------------------------")
  elif choice == 5:
    if len(queue)!=0:
      print("-------------------------Queue front {}-------------------------------".format(queue[0]))
    else:
      print("---------------------------Empty queue-----------------------------")
  elif choice == 6:
    if len(queue)!=0:
      print("---------------------------Queue front {}----------------------------".format(queue[len(queue)-1]))
    else:
      print("-----------------------------Empty queue-----------------------------")
  else:
    break

```

# Dequeue

Dequeue are group of elements that can insert, remove both from front and rear

```python
from collections import deque
D = deque()

choice = 0
while(choice != 7):
  print("1. Insert element at front")
  print("2. Insert element at rear")
  print("3. Remove element from front")
  print("4. Remove element from rear")
  print("5. Remove specific element")
  print("6. Search an element")
  print("7. Reverse the deque")
  print("8. Traverse")
  print("9. Exit")
  
  choice = int(input("--------------Enter a operation---------------- "))
  if choice == 1:
    ele = int(input("---------------Enter element---------------------"))
    D.appendleft(ele)
    print("--------------------{} inserted at position 0 --------------------".format(ele))
  elif choice == 2:
    ele = int(input("---------------Enter element---------------------"))
    D.append(ele)
    print("--------------------{} inserted at end --------------------".format(ele))
  elif choice == 3:
    D.popleft()
    print("--------------------Removed element from front --------------------")
  elif choice == 4:
    D.pop()
    print("--------------------Removed element from rear --------------------")
  elif choice == 5:
    ele = int(input("---------------Enter element---------------------"))
    if ele in D:
       D.remove(ele)
       print("--------------------Removed element {} from Deque--------------------".format(ele))
    else:
      print("Not present")
  elif choice == 6:
    ele = int(input("---------------Enter element---------------------"))
    if ele in D:
      print("Present at position {}".format(D.index(ele)))
    else:
      print("Not Present")
  elif choice == 7:
    D.reverse()
    print("Reveresed",D)
  elif choice == 8:
    for x in D:
      print(x, end =' ')
  else:
    break




```

# Infix to Postfix Conversion

```python
def Priority(x):
  if x == "+" or x=='-':
    return 1
  else:
    return 2
  
def Postfix(stri):
  pf = ""
  stack=[]
  for x in stri:
    if x == "+" or x == "-" or x=="*" or x=="/":
      if len(stack) == 0:
        stack.append(x)
      elif Priority(stack[-1]) < Priority(x):
        stack.append(x)
      else:
        while len(stack)!=0 and Priority(stack[-1])>=Priority(x):
          a = stack.pop()
          pf+=a
        stack.append(x)
    else:
      pf+=x

  while len(stack)!=0:
    pf+=stack.pop()

  return pf


stri = input("Enter an infix string")
print(Postfix(stri))
```
