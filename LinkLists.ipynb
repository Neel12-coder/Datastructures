## Linked List
Group of nodes linked together in the form of a chain

Can grow or shrink dynamically

```python
linked_list = []
choice = 0
while choice!=9:
  print("1. Add element at a position")
  print("2. Add element at the end")
  print("3. Add element at the beginning")
  print("4. Traverse linked list")
  print("5. Search element in linked list")
  print("6. Remove element")
  print("7. Remove element based on position")
  print("8. Replace an element")
  print("9. Exit")

  choice = int(input("Enter one operation "))

  if choice == 1:
    ele = input("Enter Element ")
    pos = int(input("Enter Position"))
    linked_list.insert(pos,ele)
    print(f"{ele} added at position {pos}")
  elif choice == 2:
    ele = input("Enter Element ")
    linked_list.append(ele)
    print(f"{ele} added")
  elif choice == 3:
    ele = input("Enter Element ")
    linked_list.insert(0,ele)
    print(f"{ele} inserted ")
  elif choice == 4:
    for x in linked_list:
      print(x,end=" ")
  elif choice == 5:
    ele = input("Enter Element ")
    if ele in linked_list:
      print("Present")
    else:
      print("Not Present")
  elif choice == 6:
    ele = input("Enter Element ")
    linked_list.remove(ele)
    print(f"{ele} removed")
  elif choice == 7:
    pos = int(input("Enter a position"))
    linked_list.pop(pos)
    print(f"Element at position {pos} removed")
  elif choice == 8:
    pos = int(input("Enter position "))
    ele = input("Enter element to be added")
    linked_list[pos] = ele
    print(f"{ele} updated at position {pos}")
    print(linked_list)
  else:
    print("Bye bye ")







```
